**Aim : Link Analysis and PageRank**
*   Implement the PageRank algorithm to rank web pages based on link analysis.
*   Apply the PageRank algorithm to a small web graph and analyze the results.


```python
import re

print("T074 Kermeen")
# -------------------------------
# Small Web Graph (android.com)
# -------------------------------
pages = {
    "android.com": ["developer.android.com", "source.android.com"],
    "developer.android.com": ["android.com", "source.android.com"],
    "source.android.com": ["android.com"]
}

# -------------------------------
# PageRank Implementation
# -------------------------------
def pagerank(graph, iterations=5, d=0.85):
    ranks = {page: 1/len(graph) for page in graph}

    for _ in range(iterations):
        new_ranks = {}
        for page in graph:
            rank_sum = 0
            for other_page in graph:
                if page in graph[other_page]:
                    rank_sum += ranks[other_page] / len(graph[other_page])
            new_ranks[page] = (1-d)/len(graph) + d * rank_sum
        ranks = new_ranks
    return ranks

# -------------------------------
# Run PageRank
# -------------------------------
ranks = pagerank(pages)

# -------------------------------
# Display Result
# -------------------------------
for page, score in ranks.items():
    print(page, ":", round(score, 4))
```
