A. Develop a spelling correction module using edit distance algorithms.

```python
print("T074 Kermeen")

def editDistRec(s1, s2, m, n):
    # If first string is empty
    if m == 0:
        return n
    # If second string is empty
    if n == 0:
        return m
    # If last characters of two strings are same
    if s1[m - 1] == s2[n - 1]:
        return editDistRec(s1, s2, m - 1, n - 1)
    # If last characters are not same,
    return 1 + min(editDistRec(s1, s2, m, n - 1),
                   editDistRec(s1, s2, m - 1, n),
                   editDistRec(s1, s2, m - 1, n - 1))
def editDistance(s1, s2):
    return editDistRec(s1, s2, len(s1), len(s2))
if __name__ == "__main__":
    s1 = "computer"
    s2 = "science"
    print("String 1:",s1)
    print("String 2:",s2)
    print("Edit Distance is ",editDistance(s1, s2))
```

B. Integrate the spelling correction module into an information retrieval system.

```python
print("T074 Kermeen")

def edit_distance(a, b):
    dp = [[i + j if i * j == 0 else 0 for j in range(len(b)+1)] for i in range(len(a)+1)]
    for i in range(1, len(a)+1):
        for j in range(1, len(b)+1):
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + (a[i-1] != b[j-1])
            )
    return dp[-1][-1]

dictionary = ["sky", "blue", "bright", "sun", "computer", "science"]

def correct(word):
    return min(dictionary, key=lambda w: edit_distance(word, w))

documents = [
    "The sky is blue",
    "The sun is bright",
    "Computer science is amazing"
]

query = input("Enter query: ")

corrected = correct(query)
print("Corrected word:", corrected)

print("\nSearch Results:")
for doc in documents:
    if corrected.lower() in doc.lower():
        print("-", doc)
```
