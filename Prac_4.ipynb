A. Calculate precision, recall, and F-measure for a given set of retrieval results.

```python
print("T074 Kermeen")

def calculate_metrics(retrieved_set, relevant_set):
    # True Positive (TP): items correctly retrieved
    true_positive = len(retrieved_set.intersection(relevant_set))

    # False Positive (FP): retrieved but NOT relevant
    false_positive = len(retrieved_set.difference(relevant_set))

    # False Negative (FN): relevant but NOT retrieved
    false_negative = len(relevant_set.difference(retrieved_set))

    print("True Positive:", true_positive,
          "\nFalse Positive:", false_positive,
          "\nFalse Negative:", false_negative, "\n")

    # Precision = TP / (TP + FP)
    precision = true_positive / (true_positive + false_positive)

    # Recall = TP / (TP + FN)
    recall = true_positive / (true_positive + false_negative)

    # F-measure = harmonic mean of P and R
    f_measure = 2 * precision * recall / (precision + recall)

    return precision, recall, f_measure


# Example Sets
retrieved_set = set(["doc1", "doc2", "doc3"])  # System retrieved 3 documents
relevant_set = set(["doc1", "doc4"])           # Only doc1 is truly relevant

precision, recall, f_measure = calculate_metrics(retrieved_set, relevant_set)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F-measure: {f_measure}")
```

B. Use an evaluation toolkit to measure average precision and other evaluation metrics.

```python
print("T074 Kermeen")

from sklearn.metrics import average_precision_score

# Binary Ground Truth (1 = relevant, 0 = not relevant)
y_true = [0, 1, 1, 0, 1, 1]

# Model output scores (higher score = more likely to be relevant)
y_scores = [0.1, 0.4, 0.35, 0.8, 0.65, 0.9]

# Calculate the Average Precision (AP)
average_precision = average_precision_score(y_true, y_scores)

# Print the AP score
print(f'Average precision-recall score: {average_precision}')
```

Write a program to calculate precision, recall and F-measure where true
positive is 60, false positive is 30 and false negative is 20.

Precision =Precision = TP / TP+FP

```python
print("T074 Kermeen")

# Given values
TP = 60
FP = 30
FN = 20

# Calculate Precision
precision = TP / (TP + FP)

# Calculate Recall
recall = TP / (TP + FN)

# Calculate F-measure
f_measure = 2 * (precision * recall) / (precision + recall)

# Print results
print("Precision:", precision)
print("Recall:", recall)
print("F-measure:", f_measure)
```
