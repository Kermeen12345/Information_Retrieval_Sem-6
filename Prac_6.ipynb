*   Apply the clustering algorithm to a set of documents and evaluate the clustering results.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = [
    "Cats are known for their agility and grace",          # cat doc 1
    "Dogs are often called man's best friend",             # dog doc 1
    "Some dogs are trained to assist people with disabilities",  # dog doc 2
    "The sun rises in the east and sets in the west",       # sun doc 1
    "Many cats enjoy climbing trees and chasing toys"      # cat doc 2
]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)
#KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
#print cluster
print("T074 Kermeen")
print("Cluster labels:")
for i, label in enumerate(kmeans.labels_):
    print(f"Document {i+1}: Cluster {label}")
```
